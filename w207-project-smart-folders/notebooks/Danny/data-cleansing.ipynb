```python
import pandas as pd
import re
```

```python
raw_emails = pd.read_csv('../source-data/emails.csv')
raw_emails.head()
```

```python
def split_body_from_headers(msg):
    return (msg[:-(len(msg)-(msg.index('\n\n')))], msg[(msg.index('\n\n')+2):])
```

```python
new_df = raw_emails['message'].apply(lambda x: pd.Series(split_body_from_headers(x)))
```

```python
new_df.head()
```

```python
def extract_headers(text):
    headers = re.split(r'^|\n[^:]*:', text)
    try:
        headers = [x.replace('Message-ID:', '').lstrip().rstrip() for x in headers]
    except ValueError:
        print repr(text)
        print repr(headers)
        raise ValueError
    return pd.Series(headers)

extract_headers(new_df.loc[0, 0])
```

```python
header_df = new_df.loc[:,0].apply(lambda x: extract_headers(x))
```

```python
header_df.head(10)
```

```python
header_labels = [
    'Message-ID', 
    'Date', 
    'From', 
    'To', 
    'Subject', 
    'Mime-Version', 
    'Content-Type', 
    'Content-Transfer-Encoding', 
    'X-From', 
    'X-To', 
    'X-cc', 
    'X-bcc', 
    'X-Folder', 
    'X-Origin', 
    'X-FileName'
]
```
